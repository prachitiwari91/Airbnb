# Evaluating the New York Market for Short-Term Rental Investments

## Imports

```
%sql
CREATE WIDGET TEXT database DEFAULT "hive_metastore.sample_db"
```

```
%sql
select number_of_reviews, avg(price) as avg_price from ${database}.ny_airbnb group by all
```

```
%sql
select neighbourhood_group, avg(price) as avg_price from ${database}.ny_airbnb group by all
```

```
%sql
select neighbourhood_group, sum(number_of_reviews) as total_reviews, avg(reviews_per_month) as avg_reviews_per_month,  sum(calculated_host_listings_count) as total_listing from ${database}.ny_airbnb group by all
```

```
%sql
select room_type,avg(price) as avg_price, sum(calculated_host_listings_count) as total_listings from ${database}.ny_airbnb group by all
```

```
%sql
select room_type,avg(price) as avg_price,sum(number_of_reviews),neighbourhood_group from ${database}.ny_airbnb group by all
```

```
%sql
select minimum_nights,neighbourhood_group,sum(number_of_reviews) from ${database}.ny_airbnb group by all
```

```
%sql
select neighbourhood_group,sum(number_of_reviews) as total_reviews,avg(availability_365) as availability
from ${database}.ny_airbnb
group by all
order by 1 desc
```

#Property Sales Data Analysis

```
%sql
select year(sale_date),month(sale_date),BOROUGH_NAME,count(*),avg(SALE_PRICE) from ${database}.ny_property_sales group by all
```

```
%sql
SELECT  BOROUGH_NAME,avg(SALE_PRICE),avg(price) from ${database}.ny_property_sales join ${database}.ny_airbnb on BOROUGH_NAME = neighbourhood_group and lower(NEIGHBORHOOD)= lower(neighbourhood) group by all
```

```
%sql
SELECT  NEIGHBORHOOD,neighbourhood_group,avg(SALE_PRICE) as avg_sale_price,avg(price) as avg_rental_price,sum(number_of_reviews) from ${database}.ny_property_sales join ${database}.ny_airbnb on BOROUGH_NAME = neighbourhood_group and lower(NEIGHBORHOOD)= lower(neighbourhood) group by all
```
